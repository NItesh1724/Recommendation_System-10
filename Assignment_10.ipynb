```python
user_book_matrix=pd.pivot_table(data=book,columns='Book_title',index='Userid',values='Rating')
user_book_matrix
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(strategy='mean')
```

```python
user_book_matrix_filled=pd.DataFrame(imputer.fit_transform(user_book_matrix),columns=user_book_matrix.columns,index=user_book_matrix.index)
user_book_matrix_filled
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_sim_mat=cosine_similarity(user_book_matrix_filled)
```

```python
user_sim_mat
```

```python
np.fill_diagonal(user_sim_mat,0)
user_sim_mat
```

```python
user_sim_df=pd.DataFrame(user_sim_mat,columns=book['Userid'].unique(),index=book['Userid'].unique())
user_sim_df
```
